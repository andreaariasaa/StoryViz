```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data into dataframes
main_df = pd.read_csv('Data/main_table.csv', index_col=0)
hh_df = pd.read_csv('Data/hh_roster.csv', index_col=0)
ext_df = pd.read_csv('Data/mig_ext_roster.csv', index_col=0)
int_df = pd.read_csv('Data/mig_int_roster.csv', index_col=0)
pend_df = pd.read_csv('Data/mig_pend_roster.csv', index_col=0)
lookup = pd.read_excel('Data/survey data look-up table.xlsx', sheet_name='answer_lookup')
main_type = pd.read_excel('Data/survey data look-up table.xlsx', sheet_name='main_table')

main_type = pd.read_excel('Data/survey data look-up table.xlsx', sheet_name='main_table')
for type in ['hh','mig_ext', 'mig_int', 'mig_pend']:
    append_type = pd.read_excel('Data/survey data look-up table.xlsx', sheet_name=f'{type}_roster')
    main_type = main_type.append(append_type)
```

```python
# main_df.iloc[:,0:0]
# main_df.iloc[:]['income_satisfaction']
filtered_df = narrow_data(main_df, 'income_satisfaction', 'We are in a very critical situation with current income')

new_df = []
new_df = filtered_df[['country','mig_ext_intention', 'mig_ext_plan_12months', 'nivel_violencia']].copy()
countries = {
    'SLV': "El Salvador",
    'GT': 'Guatemala',
    'HND': 'Honduras'
}

new_df.loc[new_df['country'] == 'SLV', 'country'] = 'El Salvador'
new_df.loc[new_df['country'] == 'GT', 'country'] = 'Guatemala'
new_df.loc[new_df['country'] == 'HND', 'country'] = 'Honduras'



question = 'mig_ext_intention'
question_dict = {}

for i in range(len(new_df[question])):
    result = new_df.iloc[i][question]
    question_type = main_type.loc[main_type['label'] == question]['type'].values[0]
    if question_type == 'c':
        look_key = question
    else:
        look_key = question_type
    new_lookup = lookup[np.logical_and(lookup['label'] == look_key, lookup['name'] == result)]
    result_code = new_lookup.iloc[0]['text_content']
    question_dict[result] = result_code
    
    
question_dict[99] = "No Response"
for result in question_dict:
    new_df.loc[new_df[question] == result, question] = question_dict[result]

    
question = 'mig_ext_plan_12months'
question_dict = {}

for i in range(len(new_df[question])):
    result = new_df.iloc[i][question]
    if (result == 1 or result == 0 or result == 99):
        question_type = main_type.loc[main_type['label'] == question]['type'].values[0]
        if question_type == 'c':
            look_key = question
        else:
            look_key = question_type
        new_lookup = lookup[np.logical_and(lookup['label'] == look_key, lookup['name'] == result)]
        result_code = new_lookup.iloc[0]['text_content']
        question_dict[result] = result_code
    
    
question_dict[99] = "No Response"
question_dict[''] = 'Not Asked'
new_df.mig_ext_plan_12months = new_df.mig_ext_plan_12months.fillna('')
for result in question_dict:
    new_df.loc[new_df[question] == result, question] = question_dict[result]
    
    
question = 'nivel_violencia'
question_dict = {}
new_df.nivel_violencia = new_df.nivel_violencia.fillna('')


for i in range(len(new_df[question])):
    result = new_df.iloc[i][question]
    
    if (result != ''):
        question_type = main_type.loc[main_type['label'] == question]['type'].values[0]
        if question_type == 'c':
            look_key = question
        else:
            look_key = question_type
        new_lookup = lookup[np.logical_and(lookup['label'] == look_key, lookup['name'] == result)]
        result_code = new_lookup.iloc[0]['text_content']
        question_dict[result] = result_code
    
    
question_dict[99] = "No Response"
question_dict[''] = 'Not Asked'
for result in question_dict:
    new_df.loc[new_df[question] == result, question] = question_dict[result]


new_df.to_csv('next_main_df.csv')
```
